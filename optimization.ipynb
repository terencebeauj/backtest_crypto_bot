```python
import import_ipynb
import pandas as pd
from backtesting import Backtest
import skopt
from class_strategy import Ichimoku_BB
```

```python
candles = pd.read_csv("datas/BTCUSDT_15m_2020-01-01_2021-01-01.csv")
candles.index = pd.to_datetime(candles.Opentime)
candles.drop("Opentime", inplace = True, axis = 1)
```

```python
%%time

bt = Backtest(candles, Ichimoku_BB, cash=100000, commission = 0.001)
stats = bt.optimize(mu1_ichi = [5,10,15,20,25,30],
                    mu2_ichi = [20,25,30,35,40,45,50],
                    mu3_ichi = [40,45,50,55,60,65,70],
                    mu_bb = [10,15,20,25,30,35,40,45,50],
                    sigma_bb = [1,2,3,4],
                    maximize = "Sharpe Ratio",
                    constraint = lambda param: param.mu1_ichi < param.mu2_ichi < param.mu3_ichi,
                    method = "skopt"
                    )

stats
```
